Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK.

Matplotlib is used for creating static, animated, and interactive visualizations in Python. It can be used to create a wide variety of plots and visualizations.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot: plt.plot()
Bar chart: plt.bar()
Histogram: plt.hist()
Scatter plot: plt.scatter()
Pie chart: plt.pie()


```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Plot 4')

# Display the figure with subplots
plt.tight_layout()
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot

# A bar plot is a type of chart that represents categorical data with rectangular bars. Each bar has a height corresponding to the value it represents. Bar plots are used when we want to compare a single category of data between individual sub-items.

# Here's how you can plot a bar plot and a horizontal bar plot with the provided data:

import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()

# A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. It can provide a visual summary of many important aspects of a distribution, such as the central value, variability, and skewness.

# Here's how you can plot a box plot with the provided data:

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=['Box1', 'Box2'])
plt.title('Box Plot')
plt.show()
```
